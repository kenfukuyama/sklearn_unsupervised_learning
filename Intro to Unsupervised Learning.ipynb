# Introduction
In this lecture, we have learned how to cluster data to use for unsupervised learning.

### Imports
Import libraries and write settings here.

```python
# Data manipulation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
```

### Using make_blobs from sklearn.datasets to Make a Random Cluster.

```python
# 2D Cluster.
C = [[1, 1], [-1, -1], [1, -1], [-1, 1]] # Those are the centers of cluster.
"""The centers' dimension will override the n_features."""
X, y = make_blobs(n_samples=500, centers=C, cluster_std=0.2, n_features=2, random_state=42) # Making blob. 
plt.scatter(X[:, 0], X[:, 1]);
```

```python
# 5D Cluster
C2 = [[4, 7, 5, 6, 2], [-1, 2, -1, 3, -1], [0, 0, 0, 0, 0]] 
X2, y2 = make_blobs(n_samples=500, centers=C2, cluster_std=0.2, random_state=42) # n_features =  5 is unnecessary.
plt.scatter(X2[:, -2], X2[:, -1]); # The last two numbers of the ceter array are the new centers.
```

```python
""" Very Interesting to see. X[0] = X[0] ploting that's why it's linear. i.e. around 0. 0.5 v 0.5 or 0.9 v 0.9 is plotted."""
plt.scatter(X2[:, 0], X2[:, 0]);
```

### Using DBSCAN from sklearn.cluster to Group Clusters Together.

```python
# Using DBSCAN for 2D.
db = DBSCAN().fit(X)
# db.labels_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=db.labels_);
```

```python
# Using DBSCAN for 5D.
db2 = DBSCAN().fit(X2)
# db2.labels_
```

```python
plt.scatter(X2[:, -2], X2[:, -1], c=db2.labels_);
```

```python
# Making subplots using plt.subplot method.
plt.subplot(2, 2, 1)
plt.scatter(X2[:, -2], X2[:, -1], c=db2.labels_);
plt.subplot(2, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=db.labels_);
plt.subplot(2, 2, 3)
plt.scatter(X[:, 0], X[:, 1], c=db.labels_);
plt.subplot(2, 2, 4)
plt.scatter(X[:, 0], X[:, 1], c=db.labels_);
```
